```python
df_food=pd.concat([df2,df3,df4], ignore_index=True)
df_food.head()
len(df_food)
```

```python
df_food['CLAC1_NM'].unique()
```

```python
df_food['PD_AMOUNT']=df_food['PD_BUY_CT']*df_food['PD_BUY_AM']
```

```python
df_food.head()
df_food.info()
```

```python
def to_season(m):
    if 3<= m<= 5:
        return "spring"
    if 6<= m<= 8:
        return "summer"
    if 9<= m<= 11:
        return "autumn"
    if 12<= m<= 12 or 1<= m <=2:
        return "winter"
```

```python
df_food["SEASON"] = df_food["SESS_DT"].dt.month.map(to_season)
df_food.head()
```

```python
df_food['SEASON'].unique() #winter 없는지 확인, 4~9월 판매 데이터이기 때문에 winter가 있으면 안됨.
```

```python
df_food['DAY_OF_WEEK'] = df_food['SESS_DT'].dt.dayofweek
```

```python
df_food.head()
```

```python
def is_weekend(x):
    if 0<= x<= 4:
        return False
    else:
       return True
```

```python
df_food['WEEKEND'] =df_food['DAY_OF_WEEK'].apply( is_weekend)
df_food.head()
```
